```python
import os
import numpy as np


file_path = '/content/drive/MyDrive/Avengers/Avengers.txt'                      #   .

def read_file(file_path):                                                       #     content 
    with open(file_path, 'r') as file:
        content = file.readlines()
    return content                                                              # content 

def split_string(content):                                                      # 
    lines = [line.strip() for line in  content]
    return lines

def split_space(content) :                                                      #  
    space_lines = []
    for line in content:
        words = line.split(' ')
        space_lines.extend(words)
    space_lines = [item for item in space_lines if item]  #   
    return space_lines

def lowercase_strings(lines):                                                   #  
    lowercase_lines = [line.lower() for line in lines]
    return lowercase_lines

def remove_symbols(text):                                                       #  
    symbols = "~`!@#$%^&*()-_+={}[]|\\:;'<>?,./\""
    for symbol in symbols:
        text = text.replace(symbol, "")
    return text

def get_2gram(text):                                                            #2-gram 
    ngrams = []
    for i in range(len(text) - 1):
        ngrams.append(text[i:i+2])
    return ngrams



content = read_file(file_path)
lines = split_string(content)
lowercase_lines = lowercase_strings(lines)
space_lines = split_space(lowercase_lines)

#   2-gram  
processed_lines = []
for line in space_lines:
    line = remove_symbols(line)
    ngrams = get_2gram(line)
    processed_lines.extend(ngrams)

print(processed_lines)

from collections import Counter

# processed_lines 2-gram   
ngram_counter = Counter(processed_lines)

#    2-gram  
most_common_pair = ngram_counter.most_common(1)

print(most_common_pair)
```
